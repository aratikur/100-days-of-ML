```python
import requests
import pandas as pd
```

```python
all_dfs = []  
```

```python
for i in range(1, 115):
    response = requests.get(
        f'https://api.themoviedb.org/3/tv/top_rated?api_key=62c0dac31cb341c78de8caaffa2c164f&language=en-US&page={i}'
    )
    temp_df = pd.DataFrame(response.json()['results'])[
        ['id','name','overview','first_air_date','popularity','vote_average','vote_count']
    ]
    all_dfs.append(temp_df)  

df = pd.concat(all_dfs, ignore_index=True)
```
